```python
##Choosing which pairwise columns to drop based on importance 
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) >= threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
correlated_features = correlation(data, 0.5)
correlated_features
```

```python
#Split dataset into inputs and the output (target)
X = data.iloc[:,0:-1]

#shares column- multiclass: Should be binary to classify as it is the target attribute. 
bins = [0, data['num_shares'].mean(), data['num_shares'].max()+1]
data["num_shares"]=pd.cut(data['num_shares'], bins,labels=[0, 1])
Y = data["num_shares"] 
```

```python
#import decision tree library
from sklearn.tree import DecisionTreeClassifier
```

```python
#Split train and test sets (lower overfitting)
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X, Y, random_state=42)
#sfs CANNOT REMOVE FEATURES ONCE ADDED TO INCREASE THE ACCURACY (EFS is better for that but it has a slow computution time)
sfs1 = SFS(DecisionTreeClassifier(max_depth=7),
          k_features = (1, 9),   
          forward= True,
          verbose= 2,
          scoring= 'accuracy',
          cv = 4,                
          n_jobs= -1
).fit(X_train1, Y_train1)
```
